```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# --- Step 1: Import SoothSkiSkiData.xlsx as a table ---
# Make sure the file is in the same directory or provide the full path
file_path = "SoothSkiSkiData.xlsx"
df = pd.read_excel(file_path)
df = df.T
# Optional: print the first few rows to confirm data loaded correctly
#print("Data preview:")
#print(df.head())


# --- Step 2: Turn the table into a numpy array where each row is a feature and each column is a sample ---
# By default, DataFrame has rows = samples, columns = features.
# So we transpose to make rows = features, columns = samples.
data = df.to_numpy()

data = data[2:, [1, 8, 7, 10, 4]]
for i in range(0,data.shape[0]):

    data[i,4] = float(data[i,4][0:4])


print(f"Data shape (features x samples): {data.shape}")

# # --- Step 3: Run PCA on the array ---
# # sklearn expects shape (samples, features), so transpose back temporarily
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data)

print("Explained variance ratio:", pca.explained_variance_ratio_)

# --- Step 4: Plot the first two principal components of the data ---
plt.figure(figsize=(8,6))
plt.scatter(principal_components[:, 0], principal_components[:, 1], alpha=0.7, label="Data Samples")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA of SoothSki Data")

# --- Step 5: Plot an additional feature vector on top as a star ---
# Example feature vector (replace with your own)
feature_vector = np.random.rand(data.shape[0])  # one feature across samples
# Project it onto the same PCA space
feature_vector_pca = pca.transform(np.array([199.99, 60.29796, 092.38286, 28.627, 157.8132]).reshape(1,-1))

plt.scatter(
    feature_vector_pca[0, 0],
    feature_vector_pca[0, 1],
    color='red',
    marker='*',
    s=200,
    label='Optimized Ski Vector'
)

plt.legend()
plt.grid(True)
plt.show()
```
